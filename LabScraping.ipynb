```python
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
print(soup.prettify())
```

```python
soup.select('#chart-position-1 > div.chart-content.col-xs-12.col-sm-8 > p > cite')
```

```python
soup.select("p cite") # all the song titles
```

```python
# get all artists

#chart-position-1 > div.chart-content.col-xs-12.col-sm-8 > p > em
print(soup.select('#chart-position-1 > div.chart-content.col-xs-12.col-sm-8 > p > em'))
soup.select("p em")
```

```python
#get all genre's

genre_list = soup.select('div.chart-content.col-xs-12.col-sm-8 > ul > li > a')
genre_list
```

```python
genre_list
genre_clean = []
drop = '<a href="top-new-songs.php"><em>New Release</em></a>'

for i in genre_list:
        if i in drop:
            drop(i)
        else: append.genre_clean
return genre_clean 
```

```python
#chart-position-1 > div.chart-content.col-xs-12.col-sm-8
```

```python
#initialize empty lists
title = []
artist = []
genre = []

num_iter = len(soup.select("div.chart-content.col-xs-12.col-sm-8 > p > em"))

for i in range(num_iter):
    title.append(soup.select("p cite")[i].get_text())
    artist.append(soup.select("div.chart-content.col-xs-12.col-sm-8 > p > em")[i].get_text())
    genre.append(soup.select('div.chart-content.col-xs-12.col-sm-8 > ul > li > a')[i].get_text())

print(title)
print(artist)
print(genre)

```

```python
top100 = pd.DataFrame({"title":title,
                       "artist":artist,
                       "genre":genre
                      })

top100

#still need to fix the genre ref
```
